```python
import matplotlib.pyplot as plt
from cartopy.crs import EqualEarth, Geodetic

import huracanpy

```

```python
tracks = huracanpy.load(huracanpy.example_csv_file)
tracks["pressure_cat"] = huracanpy.utils.category.get_pressure_cat(
    tracks.slp, slp_units="Pa"
)
track = tracks.groupby("track_id")[0]
plt.plot(track.lon, track.lat)
huracanpy.plot.add_tag(track.lon, track.lat, track.pressure_cat.astype(int))
```

```python
tracks = huracanpy.load(tracker="ibtracs", ibtracs_online=True, ibtracs_subset="last3years")
```

```python
track_groups = tracks.groupby("name")
track = track_groups["LEE"]
ax = plt.axes(projection=EqualEarth())
plt.plot(track.lon, track.lat, transform=Geodetic())
huracanpy.plot.add_tag(track.lon, track.lat, track.usa_status, transform=Geodetic())
ax.coastlines()
ax.set_extent([-90, -0, 10, 50])
```
